```python
import utils
import h5py
import numpy as np
```

```python
# deifne input and output paths for file conversion
outdir = '/projects/anzu2965/SPI/example_converted_files'
infile = '/scratch/alpine/anzu2965/rhybrid_job_05/state00001250.vlsv' #'./state00062000.vlsv'
outfile =  'state00001250_onlyBy.h5' #'state00062000_JC_example.h5'

# convert
#utils.vlsv_to_h5(infile, outdir, outfile)

# read in converted file and parse timestep
f = h5py.File(outdir +'/' + outfile, 'r')
t = infile.split('state')[1].split('.')[0]
```

```python
# Read in data 
# positional arrays
x = np.array(f['x'])
y = np.array(f['y'])
z = np.array(f['z'])
# magentic field
Bx =  np.array(f['magnetic_field_x'])
By =  np.array(f['magnetic_field_y'])
Bz =  np.array(f['magnetic_field_z'])
# O+ velocity
Opx =  np.array(f['O_p1_velocity_x'])
Opy =  np.array(f['O_p1_velocity_y'])
Opz =  np.array(f['O_p1_velocity_z'])
# O+ density 
Op_n =  np.array(f['O_p1_number_density'])
# O2+ velocity
O2px =  np.array(f['O2_p1_velocity_x'])
O2py =  np.array(f['O2_p1_velocity_y'])
O2pz =  np.array(f['O2_p1_velocity_z'])
# O2+ density 
O2p_n =  np.array(f['O2_p1_number_density'])
# H+ velocity
Hpx =  np.array(f['H_p1_velocity_x'])
Hpy =  np.array(f['H_p1_velocity_y'])
Hpz =  np.array(f['H_p1_velocity_z'])
# H+ density 
Hp_n =  np.array(f['H_p1_number_density'])
# e- velocity
ex =  np.array(f['electron_velocity_x'])
ey =  np.array(f['electron_velocity_y'])
ez =  np.array(f['electron_velocity_z'])
# e- density 
e_n =  np.array(f['electron_number_density'])
```

```python
# make plots

plots_dir = './RHybrid_plots/Venus_example' #'./RHybrid_plots/JC_example'  #

# --------------- Magnetic field --------------- 
label = 'Magnetic Field'
units =  '[nT]'
utils.plot_3d(x, y, x, Bx/1e10, By/1e10, Bz/1e10, t, label, units, plots_dir, label.replace(' ','_') + '.png')

# --------------------- O+ --------------------- 
# Velocity
label = 'O+ Velocity'
units =  '[km/s]'
utils.plot_3d(x, y, x, Opx, Opy, Opz, t, label, units, plots_dir, label.replace(' ','_')  + '.png')
# Density
label = 'O+ Number Density'
units =  '[1/km]'
utils.plot_1d(x, y, z, Op_n, t, label, units, plots_dir, label.replace(' ','_')  + '.png')
# --------------------- O2+ --------------------- 
# Velocity
label = 'O2+ Velocity'
units =  '[km/s]'
utils.plot_3d(x, y, x, O2px, O2py, O2pz, t, label, units, plots_dir, label.replace(' ','_')  + '.png')
# Density
label = 'O2+ Number Density'
units =  '[1/km]'
utils.plot_1d(x, y, z, O2p_n, t, label, units, plots_dir, label.replace(' ','_')  + '.png')

# --------------------- H+ --------------------- 
# Velocity
label = 'H+ Velocity'
units =  '[km/s]'
utils.plot_3d(x, y, x, Hpx, Hpy, Hpz, t, label, units, plots_dir, label.replace(' ','_')  + '.png')
# Density
label = 'H+ Number Density'
units =  '[1/km]'
utils.plot_1d(x, y, z, Hp_n, t, label, units, plots_dir, label.replace(' ','_')  + '.png')

# --------------------- e- --------------------- 
# Velocity
label = 'e- Velocity'
units =  '[km/s]'
utils.plot_3d(x, y, x, ex, ey, ez, t, label, units, plots_dir, label.replace(' ','_')  + '.png')
# Density
label = 'e- Number Density'
units =  '[1/km]'
utils.plot_1d(x, y, z, e_n, t, label, units, plots_dir, label.replace(' ','_')  + '.png')
```
